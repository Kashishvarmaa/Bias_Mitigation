```python
import pandas as pd
```

```python
# Load the dataset
file_path = 'CSV/Combined_Dataset.csv'
df = pd.read_csv(file_path)
```

```python
df.head(2)
```

```python
# Drop unnecesary columns 
drop_columns = ['USN','Section' ]
df.drop(columns=drop_columns, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
le = LabelEncoder()

# Loop through each column in the DataFrame
for column in df.columns:
    df[column] = le.fit_transform(df[column])
```

```python
df
```

```python

# Save the scaled DataFrame to a CSV file
df.to_csv('StudentFeedback_changed.csv', index=False)

# Display a message confirming the save
print("Scaled DataFrame saved to 'scaled_data.csv'")
```
